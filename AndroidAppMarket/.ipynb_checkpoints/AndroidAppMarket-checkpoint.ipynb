# Android Application Market

## Prepare dataset

```python
# read dataset
import pandas as pd
apps_with_duplicates = pd.read_csv("/Users/quangtn/Desktop/01_work/01_job/02_ml/PySpark/AA_PySpark/AndroidAppMarket/datasets/apps.csv")
```

```python
# Drop duplicates from apps.csv
apps = apps_with_duplicates.drop_duplicates()
```

```python
# Print the total number of apps
print('Total number of apps in the dataset: ', apps.shape[0])
```

```python
apps.head(5)
```

## Data Cleaning

```python
# List of characters to remove
chars_to_remove = ['+', ',', '$']

# List of column names to clean
cols_to_clean = ['Installs', 'Price']

# Loop for each column in cols_to_clean
for col in cols_to_clean:
    # Loop for each char in chars_to_remove
    for char in chars_to_remove:
        # Replace the character with an empty string
        apps[col] = apps[col].apply(lambda x: x.replace(char, ''))
```

```python
# Print a summary of the apps dataframe
apps.info()
```

```python
import numpy as np

# Convert "Install" to float datatype
apps['Installs'] = apps['Installs'].astype('float')

# Convert "Price" to float datatype
apps['Price'] = apps['Price'].astype('float')
```

```python
# Checking dtypes of the apps dataframe
apps.info()
```

## Explore application categories

```python
import plotly
plotly.offline.init_notebook_mode(connected=True)
import plotly.graph_objs as go
```

```python
# Print the total number of unique categories
num_categories = len(apps['Category'].unique())
print('Number of categories: ', num_categories)
```

```python
# Count the number of apps in each 'Category'
num_apps_in_category = apps['Category'].value_counts()
print(num_apps_in_category)
```

```python
# Sort num_apps_in_category in descending order based on the count of apps
# in each category
sorted_num_apps_in_category = num_apps_in_category.sort_values(ascending=False)

data = [go.Bar(x=num_apps_in_category.index, y=num_apps_in_category.values,)]
```

```python
plotly.offline.iplot(data)
```

```python
## Distribution of application ratings

# average rating of apps
avg_app_rating = apps['Rating'].mean()
print('Average app rating: ', avg_app_rating)
```

```python
## Distribution of apps according to their ratings
data =[go.Histogram(x=apps['Rating'])]

## vertical dashed line to indicate the average app rating
layout = {'shapes': [{
                'type': 'line',
                'x0': avg_app_rating, 'y0': 0,
                'x1': avg_app_rating, 'y1': 1000,
                'line': {'dash': 'dashdot'}
}]
         }

plotly.offline.iplot({'data': data, 'layout':layout})
```

```python
import seaborn as sns
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings('ignore')
```

```python
# select rows where both 'Rating' and 'Size' values are present
apps_with_size_and_rating_present = apps[apps.Rating.notnull() & apps.Size.notnull()]
```

```python
# Subset for categories at least 250 apps
large_categories = apps_with_size_and_rating_present.groupby('Category').filter(lambda x: len(x) >= 250)
```

```python
# Plot size vs rating
plt1 = sns.jointplot(x=large_categories['Size'], y=large_categories['Rating'])
```

```python
# Select apps whose 'Type' is 'Paid'
paid_apps = apps_with_size_and_rating_present[apps_with_size_and_rating_present['Type'] == 'Paid']
```

```python
# Plot price vs rating
plt2 = sns.jointplot(x = paid_apps['Price'], y=paid_apps['Rating'])
```

```python
## Relationship between app category and price.

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
fig.set_size_inches(15, 8)

# Select a few popular app categories
popular_app_cats = apps[apps.Category.isin(['GAME', 'FAMILY', 'PHOTOGRAPHY',
                'MEDICAL', 'TOOL', 'FINANCE', 'LIFESTYLE', 'BUSINESS'])]

# the price trend by plotting Price vs Category
ax = sns.stripplot(x=popular_app_cats['Price'], y=popular_app_cats['Category'],
                   jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories')

# Apps whose price is greater than 200
apps_above_200 = apps[apps['Price']>200]
apps_above_200[['Category', 'App', 'Price']]
```

```python
## Filter out "Junk" apps

# Select apps priced below $100
apps_under_100 = apps[apps.Price<100]

fig, ax= plt.subplots()
fig.set_size_inches(15, 8)

# Examine price vs category with authentic apps (apps_under_100)
ax = sns.stripplot(x='Price', y='Category', data=apps_under_100, jitter=True, linewidth=1)
ax.set_title('App pricing trend across categories after filtering for junk apps')
```

```python
## Popularity of paid apps vs free apps

# data for paid apps
trace0 = go.Box(y=apps[apps['Type']=='Paid']['Installs'],name='Paid') 

# data for free
trace1 = go.Box(y=apps[apps['Type']=='Unpaid']['Installs'],name='Free')
```

```python
layout=go.Layout(title='Number of downloads of paid vs free apps',
        yaxis=dict(title='log number of downloads', type='log', autorange=True))
```

```python
# add trace0 and trace1 to a list for plotting
data = [trace0, trace1]
plotly.offline.iplot({'data':data, 'layout': layout})
```

```python
## Sentiment analysis of user reviews

# load user_review.csv
reviews_df = pd.read_csv("/Users/quangtn/Desktop/01_work/01_job/02_ml/"
"PySpark/AA_PySpark/AndroidAppMarket/datasets/user_reviews.csv")
```

```python
# join the 2 dataframes
merged_df = pd.merge(apps, reviews_df, on='App', how='inner')

# Drop NA values from Sentiment and Review columns
merged_df = merged_df.dropna(subset=['Sentiment', 'Review'])

sns.set_style('ticks')
fig, ax = plt.subplots()
fig.set_size_inches(11, 8)

#User review sentiment polarity for paid vs free apps
ax = sns.boxplot(x='Type', y='Sentiment_Polarity', data=merged_df)
ax.set_title('Sentiment Polarity Distribution')
```
