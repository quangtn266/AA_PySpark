```python
# Create the years and durations lists
years = [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020,]
durations = [103, 101, 99, 100, 100, 95, 95, 96, 93, 90]

# Create a dictionary with the 2 lists
movie_dict = {'years': years, 'durations': durations}

# Print the dictionary
movie_dict
```

```python
## Create a DataFrame from a dictionary

import pandas as pd

# Create a dataframe from the dictionary
durations_df = pd.DataFrame(movie_dict)

# Print the dataframe
durations_df
```

```python
## A visual inspection of our data
import matplotlib.pyplot as plt
fig = plt.figure()

# Draw a line plot of release_years and durations
plt.plot(durations_df['years'], durations_df['durations'])

# Create a title
plt.title('Netflix Movie durations 2011-2020')

# show the plot
plt.show()
```

```python
## Loading the rest of the data from a CSV
# Read the csv as a DataFrame
netflix_df = pd.read_csv("./datasets/netflix_data.csv")

# Print the first 5 rows of the dataframe
netflix_df.head(5)
```

```python
# Subset the dataframe for type "Movie"
netflix_df_movies_only = netflix_df[netflix_df['type'] == 'Movie']

# Select only columns of interest
netflix_movies_col_subset = netflix_df_movies_only[['title', 'country',
                'genre', 'release_year', 'duration']]

# Print the first 5 rows of Dataframe
netflix_movies_col_subset.head(5)
```

```python
## Creating a scatter plot
# Create a figure and increase the figure size
fig = plt.figure(figsize=(12, 8))

# Create a scatter plot of duration versus year
plt.scatter(netflix_movies_col_subset['release_year'], 
            netflix_movies_col_subset['duration'])

# Create a title
plt.title("Movie Duration by year of release")

# show the plot
plt.show()
```

```python
## Digging deeper
# Filter for durations shorter than 60 mins
short_movies = netflix_movies_col_subset[netflix_movies_col_subset['duration'] < 60]

# Print the first 20 rows of short movies
short_movies.head(5)
```

```python
## Making non-features films
# Define an empty list
colors=[]

# Iterate over rows of netflix_movies_col_subset:
for row, col in netflix_movies_col_subset.iterrows():
    if col['genre'] == 'Children':
        colors.append('red')
    elif col['genre'] == 'Documentaries':
        colors.append('blue')
    elif col['genre'] == 'Stand-Up':
        colors.append('green')
    else:
        colors.append('black')

# inspect the first 10 values in ur list
colors[:10]
```

```python
# Set the figure style and initialize a new figure
plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12, 8))

# Create a scatter plot of duration versus release year
plt.scatter(netflix_movies_col_subset['release_year'],
            netflix_movies_col_subset['duration'], color=colors)

# Create a title and axis labels
plt.title('Movie duration by year of release')
plt.xlabel('Release year')
plt.ylabel('Duration (min)')

# show the plot
plt.show()
```
